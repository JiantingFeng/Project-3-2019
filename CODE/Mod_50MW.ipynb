```python
# coding = utf-8
import pandas as pd
import numpy as np
from sklearn import preprocessing
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# data_train = pd.read_csv('train_10mw_1.csv')
data_train = pd.read_csv('/Users/fengjianting/Desktop/Predict/DATA/50MW/train_50mw.csv')
```

```python
data_train
```

```python
del data_train['Time']
del data_train['Pressure']
label = np.array(data_train['ActualPower'])
del data_train['ActualPower']
```

```python
feature_list = list(data_train.columns)
features = np.array(data_train)
```

```python
# 利用随机森林法进行回归分析
from sklearn.ensemble import RandomForestRegressor
# 初始化随机森林， 用100棵树， 初始值为10
rf = RandomForestRegressor(n_estimators = 100, random_state = 2, oob_score = True)
rf.fit(features, label)
```

```python
predictions = rf.predict(features)
errors = abs(predictions - label)
print("Mean Absolute Error:",np.mean(errors))
```

```python
C_i = 50
sum = 0
n = np.size(label)
for i in range(n):
    if label[i] >= 0.3:
        sum += abs(label[i] - predictions[i])
MAE = sum/(n*C_i)
print("MAE:",MAE*100,"%")
```

```python
# 对测试集和验证集的数据处理
# data_test = pd.read_csv('test_10mw_1.csv')
# data_target = pd.read_csv('validation_10mw_1.csv')
data_test = pd.read_csv('/Users/fengjianting/Desktop/Predict/DATA/50MW/test_50mw.csv')
data_target = pd.read_csv('/Users/fengjianting/Desktop/Predict/DATA/50MW/validation_50mw.csv')
# 做与之前相同的处理
del data_target['Time']
del data_test['Time']
del data_test['Pressure']
# 将实际功率作为标签进行训练
label_test = np.array(data_target['ActualPower'])
features_test = np.array(data_test)
```

```python
predictions_test = rf.predict(features_test)
errors = abs(predictions_test - label_test)
print("Mean Absolute Error:",np.mean(errors))
C_i = 50
sum = 0
n = np.size(label_test)
for i in range(n):
    if label_test[i] >= 0.3:
        sum += abs(label_test[i] - predictions_test[i])
MAE = sum/(n*C_i)
print("MAE:",MAE*100,"%")
```
